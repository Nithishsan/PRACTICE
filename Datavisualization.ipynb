```python
df2 = px.data.tips()
df2
```

```python
px.histogram(df2,x="tip",color="sex")
```

```python
#for histogram
# X should be a continous value
# color should be category value
px.histogram(df2,x="total_bill",color="sex")
```

```python
#for pie
# values should be a continous value
# names should be category value
px.pie(df2,names="day",values="total_bill")
```

```python
px.pie(df2,names="day",values="total_bill",hole=0.3)
```
